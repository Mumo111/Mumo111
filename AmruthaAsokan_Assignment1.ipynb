```python
# BUSS 5802 – Assignment 1 (Question 2)
# Name: Amrutha Asokan
# Student ID: [Your Student ID Here]
# Date: January 22, 2026

# Import math module for mathematical operations
# Per class code: Importing standard libraries
import math

# ============================================================================
# REQUIREMENT 1: Multiply 8 to the 9th power and print in a sentence
# ============================================================================
print("=" * 60)
print("REQUIREMENT 1: 8 TO THE 9TH POWER")
print("=" * 60)

# Calculate 8 raised to the 9th power using exponentiation operator
result = 8 ** 9  # Exponentiation operation using ** operator

# Print the result in a formatted sentence
print(f"8 raised to the 9th power equals {result:,}.")
print(f"Mathematically: 8^9 = {result:,}\n")

# ============================================================================
# REQUIREMENT 2: Use one of Python's 35 special keywords in an operation
# ============================================================================
print("=" * 60)
print("REQUIREMENT 2: USING 'AND' KEYWORD IN OPERATION")
print("=" * 60)

# Define three variables for the operation
num1 = 25
num2 = 40
num3 = 55

# Use 'and' keyword to combine two comparison operations
# 'and' returns True only if both conditions are True
range_check = (num1 > 20) and (num3 < 100)  # 'and' keyword in operation

# Print the operation and result
print(f"Variables: num1 = {num1}, num2 = {num2}, num3 = {num3}")
print(f"Operation: (num1 > 20) and (num3 < 100)")
print(f"Result: {range_check}")
print(f"Explanation: {range_check} because {num1} > 20 is True AND {num3} < 100 is True\n")

# ============================================================================
# REQUIREMENT 3: Ask user for input and use in printed sentence
# ============================================================================
print("=" * 60)
print("REQUIREMENT 3: USER INPUT AND OUTPUT")
print("=" * 60)

# Ask the user for input using input() function
# Inspired by Python documentation: https://docs.python.org/3/library/functions.html#input
user_name = input("Please enter your name: ")  # Get user input

# Use the input in a printed sentence
print(f"\nHello {user_name}! Welcome to BUSS 5802 - Business Analytics.")
print(f"Your name '{user_name}' has been recorded for this assignment.\n")

# ============================================================================
# REQUIREMENT 4: Create "if" condition and test with True/False cases
# ============================================================================
print("=" * 60)
print("REQUIREMENT 4: IF CONDITION WITH TRUE/FALSE TESTS")
print("=" * 60)

def check_student_status(score, attendance):
    """
    Check if a student passes based on score and attendance.
    Returns "PASS" if score >= 50 and attendance >= 75, otherwise "FAIL".
    """
    # 'if' condition using 'and' keyword
    if score >= 50 and attendance >= 75:
        return "PASS"
    else:
        return "FAIL"

# Test Case 1: Should return "PASS" (both conditions True)
print("TEST CASE 1 (Expected: PASS):")
score1 = 85
attendance1 = 90
result1 = check_student_status(score1, attendance1)
print(f"  Input: score={score1}, attendance={attendance1}%")
print(f"  Result: {result1}")
print(f"  Explanation: PASS because {score1} >= 50 (True) AND {attendance1}% >= 75% (True)\n")

# Test Case 2: Should return "FAIL" (score condition False)
print("TEST CASE 2 (Expected: FAIL):")
score2 = 45
attendance2 = 80
result2 = check_student_status(score2, attendance2)
print(f"  Input: score={score2}, attendance={attendance2}%")
print(f"  Result: {result2}")
print(f"  Explanation: FAIL because {score2} >= 50 (False) even though {attendance2}% >= 75% (True)\n")

# Test Case 3: Should return "FAIL" (attendance condition False)
print("TEST CASE 3 (Expected: FAIL):")
score3 = 70
attendance3 = 60
result3 = check_student_status(score3, attendance3)
print(f"  Input: score={score3}, attendance={attendance3}%")
print(f"  Result: {result3}")
print(f"  Explanation: FAIL because {attendance3}% >= 75% (False) even though {score3} >= 50 (True)\n")

# Test Case 4: Should return "FAIL" (both conditions False)
print("TEST CASE 4 (Expected: FAIL):")
score4 = 40
attendance4 = 65
result4 = check_student_status(score4, attendance4)
print(f"  Input: score={score4}, attendance={attendance4}%")
print(f"  Result: {result4}")
print(f"  Explanation: FAIL because {score4} >= 50 (False) AND {attendance4}% >= 75% (False)\n")

# ============================================================================
# SUMMARY: All requirements completed
# ============================================================================
print("=" * 60)
print("SUMMARY: ALL REQUIREMENTS COMPLETED")
print("=" * 60)
print("✓ Requirement 1: 8^9 calculated and printed in sentence")
print("✓ Requirement 2: 'and' keyword used in logical operation")
print("✓ Requirement 3: User input collected and used in output")
print("✓ Requirement 4: if condition tested with True and False cases")
print("=" * 60)
print("Question 2 completed successfully! Ready for submission.")
print("=" * 60)
```
